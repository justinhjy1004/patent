```python
import neo4j 
import pandas as pd
import time
```

```python
uri = "neo4j://localhost:7687"
driver = neo4j.GraphDatabase.driver(uri, auth=("neo4j","4028052721"))
```

```python
def cite(tx, citing, being_cited, similarity):
    tx.run("MATCH (p:Patent) WHERE p.id = $citing "
           "MATCH (c:Patent) WHERE c.id = $being_cited "
           "MERGE (p)-[:CITED {sim:$similarity}]->(c)",
           citing=citing, being_cited=being_cited, similarity=similarity )
```

```python
chunksize = 10 ** 4
i = 1
for chunk in pd.read_csv("/media/jho/Gamma/most_sim/cite_sims.csv", chunksize = chunksize):
    if(i > 6770):
        chunk["patent_id"] = chunk["patent_id"].astype(str)
        chunk["citation_id"] = chunk["citation_id"].astype(str)
        start = time.time()
        with driver.session() as session:
            for index,row in chunk.iterrows():
                session.write_transaction(cite, row.patent_id, row.citation_id, row.similarity)
        end = time.time()
        print(i, end=": ")
        print(str(end-start)+"s")
    i = i+1
```
